```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd

books = pd.read_csv(f"{path}/books.csv")
books.sample(10)
```

```python
print("Dataframe Shape: ", books.shape)
print("=====================")
books.info()
```

```python
books.duplicated().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
    
    
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), ax=ax, cbar=False)

plt.xlabel("Missing Values")
plt.ylabel("Columns")

plt.show()
```

```python
import numpy as np

books["age_of_book"] = 2025 - books["published_year"]
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
```

```python
books.sample(6)
```

```python
columns_of_interest = ["age_of_book", "missing_description", "num_pages", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})

heatmap.set_title("Correlation Heatmap")

plt.show()
```

```python
books_missing = books[~ books["description"].isna() &
    ~  books["num_pages"].isna() &
    ~ books["published_year"].isna() &
    ~ books["average_rating"].isna() 
    ]
```

```python
books_missing["categories"].value_counts(ascending=False).reset_index()
```

```python
category_counts = books_missing["categories"].value_counts()

plt.figure(figsize=(50,10))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.xticks(rotation=90)  
plt.xlabel("Categories")
plt.ylabel("Count")
plt.title("Count of Missing Categories")
plt.tight_layout()
plt.show()
```

```python
books_missing["words_of_description"] = books_missing["description"].str.split().str.len()
books_missing.head(6)
```

```python
plt.figure(figsize=(14,8))
sns.histplot(data=books_missing, x="words_of_description")
plt.show()
```

```python
books_missing.loc[books_missing["words_of_description"].between(1, 4), "description"]
```

```python
books_missing.loc[books_missing["words_of_description"].between(5, 14), "description"]
```

```python
books_missing_25_words = books_missing[books_missing["words_of_description"] >= 25]
```

```python
books_missing_25_words
```

```python
books_missing_25_words["title_and_subtitle"] = (
    np.where(books_missing_25_words["subtitle"].isna(), 
             books_missing_25_words["title"],
             books_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)
            )
)
```

```python
books_missing_25_words
```

```python
books_missing_25_words["tagged_description"] = books_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
(
    books_missing_25_words
    .drop(["subtitle", "age_of_book", "missing_description", "words_of_description"], axis=1)
    .to_csv("datasets/books_cleaned.csv", index=False)
)
```
