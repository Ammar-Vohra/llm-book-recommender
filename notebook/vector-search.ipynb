```python
raw_documents = TextLoader("tagged_description.txt").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
split_documents = text_splitter.split_documents(raw_documents)
```

```python
split_documents[0]
```

```python
db_books = Chroma.from_documents(
    split_documents,
    embedding=OpenAIEmbeddings()
)
```

```python
query = "A book to teach children about history"
docs = db_books.similarity_search(query, k=10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendations(query:str, top_k:int = 10) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k=50)

    books_lst = []

    for i in range(0, len(recs)):
        books_lst += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_lst)].head(top_k)
```

```python
retrieve_semantic_recommendations(query="A book to teach children about history")
```

```python
import os
import requests
from dotenv import load_dotenv

# Load the API key from the .env file
load_dotenv()
api_key = os.getenv("HUGGINGFACE_API_KEY")

# Hugging Face Inference API endpoint for zero-shot
API_URL = "https://api-inference.huggingface.co/models/facebook/bart-large-mnli"
headers = {
    "Authorization": f"Bearer {api_key}"
}

def classify_zero_shot(text, candidate_labels):
    payload = {
        "inputs": text,
        "parameters": {"candidate_labels": candidate_labels}
    }
    response = requests.post(API_URL, headers=headers, json=payload)

    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error {response.status_code}: {response.text}")
        return None
```

```python
fiction_categories = ["Fiction", "NonFiction"]
```

```python
sequence = books.loc[books["mapped_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
```

```python
classify_zero_shot(text=sequence, candidate_labels=fiction_categories)
```

```python
import numpy as np

```

```python
max_index = np.argmax(classify_zero_shot(text=sequence, candidate_labels=fiction_categories)["scores"])
max_label = classify_zero_shot(text=sequence, candidate_labels=fiction_categories)["labels"][max_index]
max_label
```

```python
def generate_prediction(sequence, candidate_labels):
    predictions = classify_zero_shot(text=sequence, candidate_labels=fiction_categories)
    max_index = np.argmax(predictions["scores"])
    max_label = predictions["labels"][max_index]
    return max_label

```

```python
generate_prediction(sequence=sequence, candidate_labels=fiction_categories)
```

```python
from tqdm import tqdm 

actual_cats = []
predicted_cats = []

for i in tqdm(range(0, 300)):
    sequence = books.loc[books["mapped_categories"] == "Fiction", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_prediction(sequence=sequence, candidate_labels=fiction_categories)]
    actual_cats += ["Fiction"]

for i in tqdm(range(0, 300)):
    sequence = books.loc[books["mapped_categories"] == "NonFiction", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_prediction(sequence=sequence, candidate_labels=fiction_categories)]
    actual_cats += ["NonFiction"]
```

```python
predicted_df = pd.DataFrame({"actual_categories": actual_cats, "predicted_categories": predicted_cats})
predicted_df
```

```python
predicted_df["correct_predictions"] = (
    np.where(predicted_df["actual_categories"] == predicted_df["predicted_categories"], 1, 0)
)

predicted_df
```

```python
predicted_df["correct_predictions"].sum() / len(predicted_df)
```

```python
isbns = []
predicted_cats = []


missing_cats = books.loc[books["mapped_categories"].isna(), ["isbn13", "description"]].reset_index(drop=True)
```
